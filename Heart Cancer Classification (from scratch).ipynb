# Sigmoid Function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

nums = np.arange(-10, 10, step = 1)

fig, ax = plt.subplots(figsize = (5, 5))
ax.plot(nums, sigmoid(nums), c = 'b')
```

# Cost Function

```python
def cost(theta, X, Y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    Y = np.matrix(Y)

    first = np.multiply(-Y, np.log(sigmoid(X * theta.T)))
    second = np.multiply((1 - Y), np.log(1 - sigmoid(X * theta.T)))

    return np.sum(first - second) / len(X)
```

```python
df.insert(0, 'Ones', 1)
```

```python
cols = df.shape[1]
X = df.iloc[ : , 0 : cols - 1]
Y = df.iloc[ : , cols - 1 : cols]

X = np.array(X.values)
Y = np.array(Y.values)
theta = np.zeros(13)

print('X shape:', X.shape) # 303 Row * 13 Columns
print('theta shape:', theta.shape)
print('Y shape:', Y.shape) # 303 Row * 1 Column
```

```python
This_Cost = cost(theta, X, Y)
print('Cost:', '%.2f' % (This_Cost * 100) + '%') # Accuracy: 30.69%
```

# Gradient Descent Optimizer

```python
def gradient(theta, X, Y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    Y = np.matrix(Y)

    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)

    error = sigmoid(X * theta.T) - Y

    for i in range(parameters):
        term = np.multiply(error, X[ : , i])
        grad[i] = np.sum(term) / len(X)
        
    return grad
```

```python
import scipy.optimize as opt
result = opt.fmin_tnc(func = cost, x0 = theta, fprime = gradient, args = (X, Y))
```

```python
print(result)
```

```python
Cost_After_Optimize = cost(result[0], X, Y)
print('Cost After Optimize:', '%.2f' % (Cost_After_Optimize * 100) + '%') # Accuracy: 64.51%
```

```python
def predict(theta, X):
    
    probability = sigmoid(X * theta.T)
    
    return [1 if x >= 0.5 else 0 for x in probability]
```

```python
theta_minimum = np.matrix(result[0])

predictions = predict(theta_minimum, X)

correct = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions, Y)]

accuracy = '%.2f' % ((sum(map(int, correct)) / len(correct)) * 100)
print(f'Accuracy: {accuracy}%')
```
